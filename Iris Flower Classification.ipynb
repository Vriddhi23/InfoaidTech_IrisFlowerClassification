```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
url="https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
```

```python
columnnames=["sepal length", "sepal width", "petal length", "petal width","flower class"]
iris=pd.read_csv(url,names=columnnames);
```

```python
iris.head()
```

```python
print(iris.isnull())
```

```python
x=iris['sepal length']
y=iris['flower class']
plt.scatter(x,y)
plt.title('Iris Flower vs Sepal length')
plt.xlabel('Sepal length')
plt.ylabel('Iris flower')
```

```python
sns.histplot(iris)
```

```python
sns.violinplot(iris)
```

```python
x=iris.drop('flower class',axis=1)
y=iris['flower class']
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=40)
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
```

```python
y_pre=knn.predict(X_test)
```

```python
#predicting the accuracy between the ytest and y pre
print("Accuracy:", accuracy_score(Y_test, y_pre))
```
